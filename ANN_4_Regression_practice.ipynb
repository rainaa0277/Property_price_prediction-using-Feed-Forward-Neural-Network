```python
X = pd.read_csv("train.csv")
y = pd.read_csv("test.csv")
```

```python
X.head(3)
```

```python
y.head(3)
```

```python
X.dropna(axis=1, inplace=True)
```

```python
X
```

```python
X = X.drop(["Id"], axis=1)
```

```python
X.head(4)
```

```python
numeric_data = X.select_dtypes(include = [np.number])
categorical_data = X.select_dtypes(exclude = [np.number])
```

```python
numeric_data.columns
```

```python
categorical_data.columns
```

```python
dummy_encoding = pd.get_dummies(categorical_data, drop_first=True)
```

```python
new_x = pd.concat([numeric_data, dummy_encoding], axis=1)
new_x.head()
```

```python
scale = StandardScaler()
```

```python
target = new_x["SalePrice"]
scaled_df = scale.fit_transform(new_x.drop("SalePrice", axis=1))
```

```python
scaled_df
```

```python
scaled_df = pd.DataFrame(scaled_df, columns=new_x.columns.drop("SalePrice"))
```

```python
scaled_df.columns
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_df, target, test_size=0.20, random_state=50)

print(X_train.shape, "X_train")
print(X_test.shape, "X_test")
print(y_train.shape, "y_train")
print(y_test.shape, "y_test")
```

```python
model = Sequential()

model.add(Dense(units = 500, input_shape=(188,), kernel_initializer="normal", activation="relu"))
model.add(Dense(units = 250, kernel_initializer="normal", activation="relu"))
model.add(Dense(units=100, kernel_initializer="normal", activation = "relu"))
model.add(Dense(units=25, kernel_initializer="normal", activation="relu"))
model.add(Dense(units=1, kernel_initializer="normal", activation="linear"))
```

```python
model.summary()
```

```python
model.compile(loss = "mean_absolute_error", optimizer="Adam", metrics=["mean_absolute_error"])
```

```python
model.fit(X_train,y_train, validation_data=(X_test, y_test), epochs=200, batch_size=50)
```

```python
predictions = model.predict(X_test)
```

```python
predictions
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, predictions)
```
